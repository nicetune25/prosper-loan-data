## Dataset Overview

> Provide a brief overview of the dataset to be presented here.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
data=pd.read_csv('prosperLoanData.csv')
```

```python
usable_data = ['LoanOriginalAmount', 'BorrowerAPR', 'BorrowerRate','StatedMonthlyIncome', 'Term', 'ProsperRating (Alpha)', "ProsperScore",
        'EmploymentStatus','ListingCreationDate','Occupation','AvailableBankcardCredit','CreditGrade',"DebtToIncomeRatio",'Investors','LoanStatus']
available_data = data[usable_data]
```

```python
base_color = sb.color_palette()[0]
```

## Plot the Prosper Rating 


```python
plt.subplots(figsize = [14,8])
plt.subplot(1, 2, 1)
apr_bins = np.arange(available_data.BorrowerAPR.min(), available_data.BorrowerAPR.max()+0.01, 0.01)
plt.hist(data=available_data, x='BorrowerAPR', bins=apr_bins);
plt.xlabel('borrower\'s Annual Percentage Rate');
plt.title('Distribution of borrower\'s Annual Percentage Rate');
plt.ylabel('frequency');

# Plot the distribution of BorrowerRate
plt.subplot(1, 2, 2)
rate_bins = np.arange(available_data['BorrowerRate'].min(), available_data.BorrowerRate.max()+0.01, 0.01)
plt.hist(data=available_data, x='BorrowerRate', bins=rate_bins)
plt.xlabel('Borrower Rate');
plt.ylabel('frequency');
plt.title('Distribution of Borrower Rate');
```

as seen from the plot, 0.35 has more borrowers Annual Percentage Rate (APR) count than others.

# plot for the distribution between loanstatus, BorrowerAPR and EmploymentStatus

```python

plt.figure(figsize = [15, 8])

plt.subplot(3, 1, 1)
sb.barplot(data = data, x = 'LoanStatus', y='BorrowerAPR', hue = 'EmploymentStatus', palette = 'inferno_r');
plt.xticks(rotation=90)
plt.legend(loc=2, ncol = 4)
```

from the chart above, it was obsereved that most delay in payment or due date defaulters where found among not employed, part-time and retired employment status

# correlation between ProsperScore to BorrowerAPR

```python
# correlation between ProsperScore to BorrowerAPR 
plt.figure(figsize=[10,10])
plt.subplot(3,1,3)
sb.violinplot(data=available_data, x='ProsperScore', y='BorrowerAPR', color=base_color);
plt.ylim((0,0.4));
plt.xlabel("Prosper Score");
plt.ylabel("Borrower APR");
plt.title("correlation between ProsperScore) to BorrowerAPR ");

```

in the violin plot above, there is a negative correlation between having a higher score and a lower interest rate. With ProsperScore there is still a decline, but the distribution of the BorrowerRate is more distributed and the IQR is generally larger for each rating
